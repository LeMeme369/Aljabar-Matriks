Lakukan tahapan pada soal 1 dengan menggunakan tensor flow

```python
import tensorflow as tf
```

```python


# buat variabel skalar x dan simpan suatu nilai skalar
x_tf = 70
x_tf
```

```python
# tampilkan tipe data dari x
x_tf = tf.Variable(106, dtype=tf.int32)
x_tf
```

```python
# buat variabel skalar y dan simpan suatu nilai skalar
y_tf = tf.Variable(7, dtype=tf.int32)

# jumlahkan x+y
tf_sum = tf.add(x_tf, y_tf)
tf_sum
```

```python

# buat variabel bernama x_float dan simpan suatu nilai desimal
xtf_float = 70.7
xtf_float
```

```python

# tampilkan tipe data dari x_float
xtf_float = tf.Variable(10.5, dtype=tf.float16)
xtf_float
```

lakukan tahapan pada soal 1 dengan menggunakan PyTorch

```python
import torch as pt
```

```python

# buat variabel skalar x dan simpan suatu nilai skalar
x_pt = 70
x_pt
```

```python

# tampilkan tipe data dari x
x_pt = pt.tensor(105, dtype=pt.int32)
x_pt
```

```python

# buat variabel skalar y dan simpan suatu nilai skalar
y_pt = 17
y_pt
```

```python

# tampilkan tipe data dari y
y_pt = pt.tensor(15, dtype=pt.int32)
y_pt
```

```python

# jumlahkan x+y
x_pt + y_pt
```

```python
# buat variabel bernama x_float dan simpan suatu nilai desimal
xpt_float = 70.7
```

```python

# jumlahkan x_float dengan y
ptfloat_sum = pt.add(xpt_float, y_pt)
ptfloat_sum
```

```python
xpt_float = pt.tensor(10.5, dtype=pt.float16)
xpt_float
```
