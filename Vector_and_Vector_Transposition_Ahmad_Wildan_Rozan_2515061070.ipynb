lakukan transpose pada vector x dan simpan pada variabel x_t

```python
x_t = x.T
x_t
```

```python
# tampilkan ukuran dari vector x_t
x_t.shape
```

```python
# lakukan transpose pada vector y dan simpan pada variabel y_t
y = np.array([[7],
              [3],
              [9]])

y_t = y.T
y_t
```

Tensorflow

```python
import tensorflow as tf
```

```python
x = tf.constant([[3],
                 [1],
                 [6]])
x
```

```python
x_t = tf.transpose(x)
x_t
```

```python
y = tf.constant([[7],
                 [3],
                 [5]])
y
```

```python
y_t = tf.transpose(y)
y_t
```
